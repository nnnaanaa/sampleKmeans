```python
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
column1 = np.random.uniform(0.01, 2, 100).reshape(-1, 1)
column2 = np.random.randint(10, 501, size=(100, 1))
data = np.concatenate((column1, column2), axis=1) # 連続変数と離散型変数を結合して配列を作成

y_pred_rnd = KMeans(
    n_clusters=2, init="random", max_iter=1, n_init=1
).fit_predict(data)

plt.scatter(data[:, 0], data[:, 1], c=y_pred_rnd, marker="x")
```

```python
normal_data = np.random.uniform(0.01, 30, 95).reshape(-1, 1)
abnormal_data = np.random.uniform(30, 60, 5).reshape(-1, 1)
column1_abn = np.concatenate((normal_data, abnormal_data), axis=0)
column2_abn = np.random.randint(10, 501, size=(100, 1))

# 連続変数と離散型変数を結合して配列を作成
data_abn = np.concatenate((column1_abn, column2_abn), axis=1)

y_pred_rnd = KMeans(
    n_clusters=5, init="random", max_iter=1, n_init=1
).fit_predict(data_abn)

plt.scatter(data_abn[:, 0], data_abn[:, 1], c=y_pred_rnd, marker="x")
```

```python
from sklearn.ensemble import IsolationForest
import numpy as np

# サンプルデータを生成
X = np.random.normal(loc=0, scale=1, size=(100, 2))

# 異常データを追加
X[0] = [10, 10]  # 例として最初のデータを異常値に設定

# Isolation Forestを訓練
clf = IsolationForest(contamination=0.01)  # contaminationは異常度の期待値を指定します
clf.fit(X)

# 結果を予測
y_pred = clf.predict(X)

# 結果を出力
print("異常値の予測結果:")
print(y_pred)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y_pred, marker="x")
```
